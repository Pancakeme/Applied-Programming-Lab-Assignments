```python
# Magic command below to enable interactivity in the JupyterLab interface
%matplotlib ipympl
# Some basic imports that are useful
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from functools import partial
```

```python
no_of_points = np.lcm.reduce([3,4,5,6,7,8])
def polygon(no_sides, no_points):
    theta = 2*np.pi/no_sides
    points_per_side = int(no_points/no_sides)
    x = np.concatenate([np.linspace(np.cos(i*theta), np.cos((i+1)*theta), points_per_side) for i in range(no_sides)])
    y = np.concatenate([np.linspace(np.sin(i*theta), np.sin((i+1)*theta), points_per_side) for i in range(no_sides)])
    return x,y

x = [polygon(3+i,no_of_points)[0] for i in range(5)]
y = [polygon(3+i,no_of_points)[1] for i in range(5)]
```

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'r')
def init():
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    return ln,
def update(frame):
    # xdata.append(frame)
    # ydata.append(np.sin(frame))
    index = int(frame)
    x_i, x_f = x[index], x[index+1]
    y_i, y_f = y[index], y[index+1]
    r = frame - index
    xdata, ydata = morph(x_f, y_f, x_i, y_i, r)
    ln.set_data(xdata, ydata)
    return ln,
```

```python
def morph(x2, y2, x1, y1, alpha):
    xm = alpha * x2 + (1-alpha) * x1
    ym = alpha * y2 + (1-alpha) * y1
    return xm, ym

# print(f"Square: {np.shape(xs)}")
ani = FuncAnimation(fig, update, frames=np.linspace(0, 5, 10),
init_func=init, blit=True, interval=10, repeat=True)
plt.show()
```
